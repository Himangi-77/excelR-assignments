```python
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
airlines = pd.read_excel('C:/ExcelR/Assignments/Datasets/EastWestAirlines.xlsx', sheet_name='data')
airlines = airlines.rename(columns = {'Award?':'Award'})
airlines.head()
```

```python
#Normalization of data
def normal_func(i):
    return (i-i.min())/(i.max()-i.min())
```

```python
airlines_norm = normal_func(airlines.iloc[:,1:])
```

```python
airlines_norm.info()
```

```python
dendrogram = sch.dendrogram(sch.linkage(airlines_norm, method = 'ward'))
```

```python
#CREATING CLUSTERS
hc = AgglomerativeClustering(n_clusters = 4, affinity = 'euclidean', linkage='average')
```

```python
#SAVE CLUSTERS FOR CHARTS
y_hc = hc.fit_predict(airlines_norm)
Clusters = pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
airlines_norm['ClusterID'] = hc.labels_
```

```python
airlines_norm
```

```python
plt.figure(figsize=(20,7))
plt.scatter(airlines_norm['ClusterID'],airlines_norm['Balance'],c=hc.labels_)
plt.ylabel('Miles eligible for award travel')
plt.xlabel('Clusters')
plt.show()
```

```python
#KMeans Clustering
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
#Normalizing the Data
scaler = StandardScaler()
scaled_X=scaler.fit_transform(airlines.iloc[:,1:])
```

```python
#Generating the dendrogram for normalized data
dendrogram = sch.dendrogram(sch.linkage(scaled_X,method='complete'))
```

```python
#To find optimum number of cluster
#The K-Means algorithm aims to choose centroids that minimize the inertia or within-cluster sum-of-squares criterion:
```

```python
wcss=[]
for i in range(1,11):
    kMeans=KMeans(n_clusters=i,random_state=0)
    kMeans.fit(scaled_X)
    wcss.append(kMeans.inertia_)
plt.plot(range(1,11),wcss)
plt.xlabel('K in K-Means')
plt.ylabel('WCSS')
plt.title('Elbow Method')
```

```python
new_cluster = KMeans(n_clusters=6,random_state=42)
new_cluster.fit(scaled_X)
```

```python
new_cluster.labels_
```

```python
airlines_norm['K_New_Cluster_ID']=new_cluster.labels_
```

```python
airlines_norm.head(10)
```

```python
new_cluster.cluster_centers_
```

```python
airlines_norm.groupby('K_New_Cluster_ID').agg(['mean']).reset_index()
```
