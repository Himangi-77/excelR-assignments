```python
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
airlines = pd.read_excel('C:/ExcelR/Assignments/Datasets/EastWestAirlines.xlsx', sheet_name='data')
airlines = airlines.rename(columns = {'Award?':'Award'})
airlines.head()
```

```python
#Normalization of data
def normal_func(i):
    return (i-i.min())/(i.max()-i.min())
```

```python
airlines_norm = normal_func(airlines.iloc[:,1:])
```

```python
airlines_norm.info()
```

```python
dendrogram = sch.dendrogram(sch.linkage(airlines_norm, method = 'complete'))
```

```python
#CREATING CLUSTERS
hc = AgglomerativeClustering(n_clusters = 4, affinity = 'euclidean', linkage='average')
```

```python
#SAVE CLUSTERS FOR CHARTS
y_hc = hc.fit_predict(airlines_norm)
Clusters = pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
airlines_norm['ClusterID'] = hc.labels_
```

```python
airlines_norm
```

```python
plt.figure(figsize=(20,7))
plt.scatter(airlines_norm['ClusterID'],airlines_norm['Balance'],c=hc.labels_)
plt.ylabel('Miles eligible for award travel')
plt.xlabel('Clusters')
plt.show()
```
