```python
'''
Problem statement.

Build a recommender system by using cosine simillarties score.
'''
```

```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv('/content/book.csv', encoding='latin')
books.head()
```

```python
books.columns = ['ID','UserID','BookTitle','BookRating']
```

```python
books_df = books.pivot(index='ID',columns='BookTitle',values='BookRating').reset_index(drop=True)
books_df
```

```python
books_df.fillna(0,inplace=True)
books_df.head()
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1-pairwise_distances(books_df.values,metric='cosine')
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df.index = books.ID.unique()
user_sim_df.columns = books.ID.unique()
user_sim_df.iloc[0:5,0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5,0:5]
```

```python
#Most Similar Users
simi_user = user_sim_df.idxmax(axis=1)[0:20]
simi_user_df = pd.DataFrame(simi_user).reset_index()
simi_user_df.columns = ['User1ID', 'User2ID']
simi_user_df
```
