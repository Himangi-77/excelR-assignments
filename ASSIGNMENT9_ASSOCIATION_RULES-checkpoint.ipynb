```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
books = pd.read_csv('C:/ExcelR/Assignments/Datasets/book.csv')
books.head()
```

```python
books.info()
```

```python
frequent_itemsets = apriori(books, min_support = 0.1, use_colnames = True)
len(frequent_itemsets)
```

```python
frequent_itemsets1 = apriori(books, min_support = 0.2, use_colnames = True)
len(frequent_itemsets1)
```

```python
frequent_itemsets3 = apriori(books, min_support = 0.3, use_colnames = True)
len(frequent_itemsets3)
```

```python
#We observe that as the minimum support increases the number of observation decreases and vice versa. Thus, support percentage and no. of associations are inversely related.
```

```python
#We stick with min_support 0.1
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
rules
```

```python
#A leverage value of 0 indicates independence
#A high conviction value means that the consequent is highly depending on the antecedent
```

```python
rules.sort_values("confidence", ascending=False).head()
```

```python
#High confidence means strong confidence in practice
```

```python
rules[(rules.support>0.15)&(rules.confidence>0.4)].sort_values("confidence",ascending=False)
```

```python
rules['lhs items'] = rules['antecedents'].apply(lambda x:len(x) )
rules[rules['lhs items']>1].sort_values('lift', ascending=False).head()
```

```python
rules['antecedents_'] = rules['antecedents'].apply(lambda a: ','.join(list(a)))
rules['consequents_'] = rules['consequents'].apply(lambda a: ','.join(list(a)))
# Transform the DataFrame of rules into a matrix using the confidence metric
pivot = rules[rules['lhs items']>1].pivot(index = 'antecedents_', 
                    columns = 'consequents_', values= 'confidence')
# Generate a heatmap with annotations 
sns.heatmap(pivot, annot = True)
plt.title('Heat Map - For Confidence Metric')
plt.yticks(rotation=0)
plt.xticks(rotation=90)
```

```python
rules_li = association_rules(frequent_itemsets, metric = "lift", min_threshold = 0.7)
rules_li
```

```python
rules_li['lhs'] = rules_li['antecedents'].apply(lambda x: len(x))
rules_li[rules_li.lhs>1].sort_values("lift",ascending=False).head()
```

```python
rules_li['_antecedents'] = rules_li['antecedents'].apply(lambda a:','.join(list(a)))
rules_li['_consequents'] = rules_li['consequents'].apply(lambda a:','.join(list(a)))
#Transforming the data frame into a matrix using lift metric
pivot = rules_li[rules_li.lhs>1].pivot(index='_antecedents', columns='_consequents',values='lift')
#Transforming the matrix into a heatmap
sns.heatmap(pivot,annot=True)
plt.title('Heatmap for Lift Matrix')
plt.yticks(rotation=0)
plt.xticks(rotation=90)
```
