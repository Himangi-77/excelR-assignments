```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:/ExcelR/Assignments/Datasets/bank-full.csv')
data
```

```python
data[['Age','Job','Marital','Education','Default','Balance','Housing','Loan','Contact','Day','Month','Duration','Campaign','Pdays','Previous','Poutcome','Y']]=data['age;"job";"marital";"education";"default";"balance";"housing";"loan";"contact";"day";"month";"duration";"campaign";"pdays";"previous";"poutcome";"y"'].str.split(';',expand=True)
```

```python
data.drop(['age;"job";"marital";"education";"default";"balance";"housing";"loan";"contact";"day";"month";"duration";"campaign";"pdays";"previous";"poutcome";"y"'],axis=1)
```

```python
data1 = pd.get_dummies(data,columns=['Job','Marital','Education','Contact','Month','Poutcome'])
data1
```

```python
data1['Default']=np.where(data1['Default'].str.contains("yes"),1,0)
data1['Housing']=np.where(data1['Housing'].str.contains("yes"),1,0)
data1['Loan']=np.where(data1['Loan'].str.contains("yes"),1,0)
data1['Y']=np.where(data1['Y'].str.contains("yes"),1,0)
data1
```

```python
#To see all the columns
pd.set_option("display.max.columns",None)
data1
```

```python
data1.info()
```

```python
#Dividing the data into input and output variables
X=pd.concat([data1.iloc[:,0:10],data1.iloc[:,11:]],axis=1)
Y=data1.iloc[:,10]
```

```python
#MODEL BUILDING
```

```python
classifier = LogisticRegression()
classifier.fit(X,Y)
```

```python
#Predicting y for x dataset
y_pred = classifier.predict(X)
```

```python
y_pred_df = pd.DataFrame({'Actual':Y,'Predicted':y_pred})
y_pred_df
```

```python
#Confusion Matrix for the Model Accuracy
```

```python
confusion_matrix = confusion_matrix(Y,y_pred)
print(confusion_matrix)
```

```python
#Manual Testing of model accuracy: a+d/a+b+c+d
(39155+1166)/(39155+767+4123+1166)
```

```python
# As accuracy = 0.8933, which is greater than 0.5; Thus [:,1] Threshold value>0.5=1 else [:,0] Threshold value<0.5=0 
classifier.predict_proba(X)[:,1] 
```

```python
print(classification_report(Y,y_pred))
```

```python
fpr, tpr, thresholds = roc_curve(Y, classifier.predict_proba(X)[:,1])

auc = roc_auc_score(Y,y_pred)

plt.plot(fpr,tpr,color='red',label='logit model (area = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')

print("Area Under Curve Accuracy:", auc)
```
