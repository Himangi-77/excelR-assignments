```python
#CALCULATING VIF FOR VALIDATION TEST
```

```python
rsq_RnD = smf.ols('RnD_Spend~Administration+Marketing_Spend', data=startups).fit().rsquared
vif_RnD = 1/(1-rsq_RnD)

rsq_a = smf.ols('Administration~RnD_Spend+Marketing_Spend', data=startups).fit().rsquared
vif_a = 1/(1-rsq_a)

rsq_ms = smf.ols('Marketing_Spend~RnD_Spend+Administration', data=startups).fit().rsquared
vif_ms = 1/(1-rsq_ms)

d1 = {'Variables':['RnD Spend','Administration','Marketing Spend'],'VIF':[vif_RnD,vif_a,vif_ms]}
d2 = {'Variables':['RnD Spend','Administration','Marketing Spend'],'RsquaredValues':[rsq_RnD,rsq_a,rsq_ms]}
rsq_frame = pd.DataFrame(d2)
vif_frame = pd.DataFrame(d1)
```

```python
# 2) Residual Analysis
# Test for Normality of Residuals (Q-Q Plot) using residual model (model.resid)

sm.qqplot(model.resid,line='q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid<-30000))
```

```python
#Test for Homosedacity or Heterosedacity: Plotting model's standardized fitted values vs standardized residual values
def get_standardized_values(vals):
    return (vals-vals.mean())/vals.std() #User-defined z-score
```

```python
plt.scatter(get_standardized_values(model.fittedvalues),get_standardized_values(model.resid))
plt.xlabel('Standardized Fitted Values')
plt.ylabel('Standardized Residual Values')
plt.show()
```

```python
#MODEL DETECTION DIAGNOSTICS
#Detecting Influencers/Outliers
#Methods: 1.Cook's Distance 2.Leverage Value
```

```python
#Cook's Distance
(c,_) = model.get_influence().cooks_distance
c
```

```python
#Plot influencers using stem
fig = plt.figure(figsize=(20,7))
plt.stem(np.arange(len(startups)),np.round(c,3))
plt.xlabel('Row Index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
#Get the index and value of influencer where c>0.5
np.argmax(c), np.max(c)
```

```python
#Leverage Value using High-Influencer points: Points beyond leverage cutoff values are influencers
fig, ax = plt.subplots(figsize = (20,20))
fig = influence_plot(model, ax=ax)
```
