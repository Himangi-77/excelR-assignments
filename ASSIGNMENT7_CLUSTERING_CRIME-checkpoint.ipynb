```python
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
crime = pd.read_csv('C:/ExcelR/Assignments/Datasets/crime_data.csv')
crime.head()
```

```python
crime.info()
```

```python
def norm_func(i):
    return (i-i.min())/(i.max()-i.min())
```

```python
#Normalize the values
crime_norm = norm_func(crime.iloc[:,1:])
```

```python
#Generate the dendrogram
dendrogram = sch.dendrogram(sch.linkage(crime_norm, method = 'average'))
```

```python
#CREATE CLUSTER
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'complete')
```

```python
y_hc = hc.fit_predict(crime_norm)
Clusters = pd.DataFrame(y_hc, columns=['Clusters'])
Clusters
```

```python
crime_norm['HClustersID'] = hc.labels_
crime_norm
```

```python
#K-Means Clustering
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
#Normalizing the data
scaler = StandardScaler()
scaled_crime = scaler.fit_transform(crime.iloc[:,1:])
```

```python
#Generating the dendrogram on normalized data
dendrogram = sch.dendrogram(sch.linkage(crime_norm, method = 'complete'))
```

```python
#To find optimum number of cluster
#The K-Means algorithm aims to choose centroids that minimize the inertia or within-cluster sum-of-squares criterion:
```

```python
wcss=[]
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state = 0)
    kmeans.fit(scaled_crime)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('No. of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
new_cluster = KMeans(n_clusters=4,random_state = 42)
new_cluster.fit(scaled_crime)
```

```python
new_cluster.labels_
```

```python
crime['New_ClusterID'] = new_cluster.labels_
```
