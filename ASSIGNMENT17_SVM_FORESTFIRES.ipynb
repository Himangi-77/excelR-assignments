classify the Size_Categorie using SVM

month	month of the year: 'jan' to 'dec'

day	day of the week: 'mon' to 'sun'

FFMC	FFMC index from the FWI system: 18.7 to 96.20

DMC	DMC index from the FWI system: 1.1 to 291.3

DC	DC index from the FWI system: 7.9 to 860.6

ISI	ISI index from the FWI system: 0.0 to 56.10

temp	temperature in Celsius degrees: 2.2 to 33.30

RH	relative humidity in %: 15.0 to 100

wind	wind speed in km/h: 0.40 to 9.40

rain	outside rain in mm/m2 : 0.0 to 6.4

Size_Categorie 	the burned area of the forest ( Small , Large)

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
data = pd.read_csv('/content/forestfires.csv')
data.head()
```

```python
pd.set_option('display.max_columns', None)
data.head()
```

```python
le = LabelEncoder()
data['month'] = le.fit_transform(data['month'])
data['day'] = le.fit_transform(data['day'])
data['size_category'] = le.fit_transform(data['size_category'])
```

```python
x = data.iloc[:,0:30]
y = data.iloc[:,30]
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.33)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

Grid Search Hyperparameter Fine Tuning

```python
clf = SVC()
param_grid = [{'kernel':['rbf'], 'gamma':[50,5,10,0.5,0.05],'C':[15,14,13,12,11,10,0.1,0.001]}]
grid = GridSearchCV(clf, param_grid, cv=10)
grid.fit(x_train, y_train)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
clf = SVC(C=15, gamma=0.05, kernel='rbf')
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
print('Accuracy Score = ', accuracy_score(y_test,y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

```python
pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
```
